```python
import requests
from bs4 import BeautifulSoup
```

```python
file_path = '/series-images.html'  # HTML dosyasının yolu

with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()
```

```python
html_content
```

```python
soup = BeautifulSoup(html_content, 'lxml')
```

```python
diziler_listesi = soup.find_all("li", class_ = "ipc-metadata-list-summary-item")
```

```python
dizi = diziler_listesi[0]
```

```python
dizi.find_all("h3", class_ = "ipc-title__text")[0].text
```

```python
dizi.find_all("a", class_ = "ipc-title-link-wrapper")[0]["href"]
```

```python
dizi.find_all("span", class_ = "sc-43986a27-8 jHYIIK dli-title-metadata-item")[0].text
```

```python
dizi.find_all("div", class_ = "ipc-html-content-inner-div")[0].text
```

```python
isimler = []
linkler = []
tarihler = []
aciklamalar = []

for dizi in diziler_listesi:
  name = dizi.find_all("h3", class_ = "ipc-title__text")[0].text
  name = name.split(". ")
  if len(name) == 2:
    name = name[1]
  else:
    print(name)
    continue
  url = dizi.find_all("a", class_ = "ipc-title-link-wrapper")[0]["href"]
  try:
    date = dizi.find_all("span", class_ = "sc-43986a27-8 jHYIIK dli-title-metadata-item")[0].text
  except:
    date = None
  try:
    text = dizi.find_all("div", class_ = "ipc-html-content-inner-div")[0].text
  except:
    text = None

  isimler.append(name)
  linkler.append(url)
  tarihler.append(date)
  aciklamalar.append(text)
```

```python
len(isimler), len(linkler), len(tarihler), len(aciklamalar)
```

```python
import pandas as pd

deneme = pd.DataFrame(isimler, columns = ["isimler"])
```

```python
pd.concat([deneme, deneme], axis = 1)
```

```python
import pandas as pd
dataset = pd.DataFrame()

dataset = pd.concat([ dataset,
                    pd.DataFrame(isimler, columns = ["isimler"]),
                    pd.DataFrame(linkler, columns = ["linkler"]),
                    pd.DataFrame(tarihler, columns = ["tarihler"]),
                    pd.DataFrame(aciklamalar, columns = ["aciklamalar"])
                     ],
                    axis = 1)
dataset
```

```python
dataset.to_excel("dataset.xlsx", index= False)
dataset.to_csv("dataset.csv", index = False)
```
