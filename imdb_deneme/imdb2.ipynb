```python
import requests
from bs4 import BeautifulSoup

# URL'i belirtin
url = 'https://www.imdb.com/title/tt13323566/'

# İsteği yapın ve içeriği çekin
response = requests.get(url)
html_content = response.content

# Beautiful Soup objesini oluşturun
soup = BeautifulSoup(html_content, 'html.parser')

# İstediğiniz bilgileri çekin, örneğin başlık
title = soup.find('h1').text.strip()

# Sonuçları yazdırın
print("Film Başlığı:", title)
```

```python
import requests
from bs4 import BeautifulSoup

# Aranacak kelime ve URL
query = "kuş uçuşu dizisi bir bölüm kaç dakika"
url = f"https://www.google.com.tr/search?q={query}"

# Request headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}

response = requests.get(url, headers=headers)

# HTML içeriğini ayrıştır
soup = BeautifulSoup(response.text, 'html.parser')

# Sonuçları bul (örnek olarak başlıklar)
for result in soup.find_all('h3'):
    print(result.text)
```
