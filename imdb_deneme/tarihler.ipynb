```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df
```

```python
yapım_yılı = []
baslangıç_yılı = []
bitiş_yılı = []

for i in range(len(df)):
  cell = df['tarihler'][i]

  if "–" not in str(cell):
    yapım_yılı.append(cell)
    baslangıç_yılı.append(cell)
    bitiş_yılı.append(cell)

  elif "–" in cell:
    start, end = cell.split("–")

    if end == " ":
      end = start

    #print(cell, start, end)

    yapım_yılı.append(start)
    baslangıç_yılı.append(start)
    bitiş_yılı.append(end)

  else:
    print(cell)
```

```python
yapım_yılı_DF = pd.DataFrame(yapım_yılı, columns=["yapım_yılı"])
baslangıç_yılı_DF = pd.DataFrame(baslangıç_yılı, columns=["baslangıç_yılı"])
bitiş_yılı_DF = pd.DataFrame(yapım_yılı, columns=["bitiş_yılı"])

df = pd.concat([df, yapım_yılı_DF, baslangıç_yılı_DF, bitiş_yılı_DF],
               axis = 1)
```
